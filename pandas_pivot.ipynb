<h1 style="color:blue">Pivot Table</h1>

```python
df = pd.read_csv("weather2.csv")
df
```

```python
df.pivot_table(index="city",columns="date")
```

<h2 style="color:brown">Margins</h2>

```python
df.pivot_table(index="city",columns="date",margins=True,aggfunc=np.max)
```

<h2 style="color:brown">Grouper</h2>

```python
df = pd.read_csv("weather3.csv")
df
```

```python
type(df['date'][0])
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df[['humidity','temperature']][pd.DatetimeIndex(df['date']).month==12].mean()
```

```python
type(df['date'][0])
```

```python
df
```

```python
df.pivot_table(index=pd.Grouper(freq='m',key='date'),columns='city',aggfunc=np.mean)
```
