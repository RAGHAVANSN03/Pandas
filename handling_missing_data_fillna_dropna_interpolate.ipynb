## <font color="blue">fillna</font>

<font color="purple">**Fill all NaN with one specific value**</font>

```python
new_df = df.fillna(0)
new_df
```

<font color="purple">**Fill na using column names and dict**</font>

```python
new_df = df.fillna({
        'temperature':0,
        'windspeed': 99.9,
        'event': 'Event not Recorded'
    })
new_df
```

```python
new_df['temperature'].mean()
```

```python
df['temperature'].mean()
```

```python
df
```

```python
df['temperature'].fillna(0).mean()
```

```python
df
```

<font color="purple">**Use method to determine how to fill na values**</font>

```python
new_df = df.fillna(method="ffill")
new_df
```

```python
new_df['temperature'].mean()
```

```python
new_df = df.fillna(method="bfill")
new_df
```

<font color="purple">**Use of axis**</font>

```python
new_df = df.fillna(method="ffill", axis="columns") # axis is either "index" or "columns"
new_df
```

<font color="purple">**limit parameter**</font>

```python
new_df = df.fillna(method="ffill",limit=1)
new_df
```

```python
new_dft=new_df.fillna(method='bfill',limit=2)
new_dft
```

```python
new_df = df.fillna(method="bfill",limit=1)
new_df
```
